```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df = pd.read_csv('Data/shooting_wk7.csv')
```

```python


# Create the scatter plot
fig, ax = plt.subplots(figsize=(12, 8))
scatter = ax.scatter(df['xG'], df['Gls'], alpha=0.6, s=100)

# Add labels and title
ax.set_xlabel('Expected Goals (xG)', fontsize=12)
ax.set_ylabel('Actual Goals (Gls)', fontsize=12)
ax.set_title('Actual Goals vs Expected Goals', fontsize=14)

# Add a diagonal line
lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]
ax.plot(lims, lims, 'r--', alpha=0.75, zorder=0)

# Add team labels to points
for i, txt in enumerate(df['Squad']):
    ax.annotate(txt, (df['xG'].iloc[i], df['Gls'].iloc[i]), fontsize=8, alpha=0.7)

# Adjust layout to prevent clipping of tick-labels
plt.tight_layout()

# Show plot
plt.show()

print("Scatter plot of Actual Goals vs Expected Goals has been generated using matplotlib.")
```

Manchester City and Chelsea are significantly overperforming their xG.
Manchester United and Bournemouth are notably underperforming their xG.

```python

# Calculate the difference between actual goals and expected goals
df['Goals_Difference'] = df['Gls'] - df['xG']

# Sort teams by the difference
df_sorted = df.sort_values('Goals_Difference', ascending=False)

# Create a bar plot of the differences
plt.figure(figsize=(14, 10))
bars = plt.bar(df_sorted['Squad'], df_sorted['Goals_Difference'])

# Color the bars based on positive (blue) or negative (red) differences
for i, bar in enumerate(bars):
    if df_sorted['Goals_Difference'].iloc[i] < 0:
        bar.set_color('red')
    else:
        bar.set_color('blue')

plt.axhline(y=0, color='black', linestyle='-', linewidth=0.5)
plt.title('Difference between Actual Goals and Expected Goals (xG)', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Goals Difference (Actual - Expected)', fontsize=12)
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

# Add value labels on top of each bar
for i, v in enumerate(df_sorted['Goals_Difference']):
    plt.text(i, v, f'{v:.1f}', ha='center', va='bottom' if v > 0 else 'top', fontsize=9)

plt.tight_layout()
plt.show()

print("Top 5 overperforming teams:")
print(df_sorted[['Squad', 'Gls', 'xG', 'Goals_Difference']].head().to_string(index=False))
print("\
Top 5 underperforming teams:")
print(df_sorted[['Squad', 'Gls', 'xG', 'Goals_Difference']].tail().to_string(index=False))
```

```python
# 2. Shot Conversion Rate
df['Shot_Conversion'] = df['Gls'] / df['Sh'] * 100
plt.figure(figsize=(12, 6))
sns.barplot(x='Squad', y='Shot_Conversion', data=df.sort_values('Shot_Conversion', ascending=False))
plt.xticks(rotation=90)
plt.title('Shot Conversion Rate by Team')
plt.ylabel('Shot Conversion Rate (%)')
plt.tight_layout()
plt.show()
```

```python
# 3. Correlation Heatmap
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Variables')
plt.tight_layout()
plt.show()
```

```python
# 4. Top 5 teams by various metrics
metrics = ['Gls', 'Sh', 'SoT', 'xG', 'G-xG']
for metric in metrics:
    print(f"\
Top 5 teams by {metric}:")
    print(df.sort_values(metric, ascending=False)[['Squad', metric]].head().to_string(index=False))
```

```python
# 5. Distribution of shot distances
plt.figure(figsize=(10, 6))
sns.histplot(df['Dist'], kde=True)
plt.title('Distribution of Average Shot Distances')
plt.xlabel('Average Shot Distance')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
print("\
Correlation between Shot Distance and Goals:")
print(df['Dist'].corr(df['Gls']))

print("\
Correlation between Shot Distance and Expected Goals:")
print(df['Dist'].corr(df['xG']))
```

These correlations suggest that:

There is a weak negative correlation between shot distance and goals scored.
There is a moderate negative correlation between shot distance and expected goals.
This implies that teams taking shots from closer distances tend to score more goals and have higher expected goals, which is logical in football.

Overall, this EDA provides a comprehensive view of the shooting performance of Premier League teams after 7 weeks. It highlights the differences between actual and expected performance, the efficiency of teams in converting shots to goals, and the impact of shot distance on goal-scoring. Teams like Manchester City and Chelsea stand out for their overperformance, while others like Manchester United and Bournemouth are notably underperforming their expected goals.

```python
# Create target variable: 1 if overperforming (Gls > xG), 0 otherwise
df['Overperformance'] = (df['Gls'] > df['xG']).astype(int)
```

```python
features = ['Sh', 'SoT', 'SoT%', 'Sh/90', 'SoT/90', 'G/Sh', 'G/SoT', 'Dist', 'npxG']
X = df[features]
y = df['Overperformance']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

This code splits the data into training (70%) and testing (30%) sets using train_test_split. The Logistic Regression model is then trained (model.fit()) using the training data.

```python
from sklearn.metrics import classification_report

# Predict on the test set
y_pred = model.predict(X_test)

# Evaluate model performance
report = classification_report(y_test, y_pred)
print(report)
```

The model performs better at predicting underperformance (Class 0) but is very poor at predicting overperformance (Class 1).
